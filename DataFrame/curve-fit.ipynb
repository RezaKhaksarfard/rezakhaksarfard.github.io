# curve_fit

```python
import numpy as np
import random
from scipy.optimize import curve_fit

def func( x, a, b, c ):
    return a + b * np.log10( c + x )

xData = [ i for i in range( 1, 20 ) ]

# Generate yData by func( x, 3, 2, 4 ) plus noise
random.seed( 59 )
yData = [ func( x, 3, 2, 4 ) + random.uniform( -0.01, 0.01 ) for x in xData ]

# Fit data
popt, pcov = curve_fit( func, xData, yData, bounds = ( [ 0, 0, 1 ], [ 10, 10, 10 ] ) )
yFit = [ func( x, popt[ 0 ], popt[ 1 ], popt[ 2 ] ) for x in xData ]

print( 'a = ', popt[ 0 ] )
print( 'b = ', popt[ 1 ] )
print( 'c = ', popt[ 2 ] )
```

### Plot data and fit

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure( figsize = ( 8, 6 ) )
plt.scatter( xData, yData, color = 'black', label = 'Data' )
plt.plot( xData, yFit, color = 'red', label = 'Fit' )
plt.grid( linestyle = '--')
plt.xlim( 1, 21 )
plt.ylim( 4, 6 )
plt.legend( loc = 'upper left' )
plt.show()
```
