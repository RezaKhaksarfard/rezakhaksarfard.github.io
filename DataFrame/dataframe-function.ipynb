# Dataframes are not copied when passed to a function

```python
import pandas as pd

def func( df ):
    df.drop( columns = ['B'], inplace = True )

df = pd.DataFrame( { 'A': [ 'a', 'b', 'd', 'e', 'c', 'g', 'f', 'h'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )
print( 'Before calling the function')
display( df )

func( df )

print( 'After calling the function')
display( df )
```

```python
def func( df1 ):
    df2 = df1
    df2.drop( columns = ['B'], inplace = True )

df1 = pd.DataFrame( { 'A': [ 'a', 'b', 'd', 'e', 'c', 'g', 'f', 'h'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )
print( 'Before calling the function')
display( df1 )

func( df1 )

print( 'After calling the function')
display( df1 )
```

```python
def func( df1 ):
    df2 = df1.copy() # A new object is created
    df2.drop( columns = ['B'], inplace = True )

df1 = pd.DataFrame( { 'A': [ 'a', 'b', 'd', 'e', 'c', 'g', 'f', 'h'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )
print( 'Before calling the function')
display( df1 )

func( df1 )

print( 'After calling the function')
display( df1 )
```

```python
def func( df ):
    df = df[ [ 'A', 'C' ] ] # A new object is created

df = pd.DataFrame( { 'A': [ 'a', 'b', 'd', 'e', 'c', 'g', 'f', 'h'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )
print( 'Before calling the function')
display( df )

func( df )

print( 'After calling the function')
display( df )
```
