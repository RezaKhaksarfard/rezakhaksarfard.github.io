# groupby filter

```python
import pandas as pd

df = pd.DataFrame( { 'A': [ 'a', 'a', 'b', 'a', 'b', 'b', 'a', 'b'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )

print( df )
```

### Filter a group if max of column 'C' is higher than 6

```python
def func( group ):
    return group[ 'C' ].max() > 6

print( df.groupby( by = 'A' ).filter( func ))
```

```python
def func( group ):
    return group[ 'C' ].max() > 6

print( df.groupby( by = 'A' ).filter( func ).reset_index( drop = True ) )
```
