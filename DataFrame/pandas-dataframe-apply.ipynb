# Add a new column to dataframe by pandas.DataFrame.apply 

```python
import pandas as pd

dfSource = pd.DataFrame( { 'A': [ 'a', 'a', 'b', 'a', 'b', 'b', 'a', 'b'],
                           'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                           'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )

display( dfSource )
```

```python
def func( row ):
    if row[ 'B' ] > 4:
        return row[ 'B' ] - row[ 'C' ]
    else:
        return row[ 'B' ] + row[ 'C' ]

df = dfSource.copy()
df[ 'D' ] = df.apply( func, axis = 1 )
display( df )
```

```python
# function with extra parameter

def func( row, x ):
    if row[ 'B' ] > 4:
        return row[ 'B' ] - row[ 'C' ] * x
    else:
        return row[ 'B' ] + row[ 'C' ]

    return row[ 'B' ] + row[ 'C' ] * x

df = dfSource.copy()

x = 2
df[ 'D' ] = df.apply( lambda row: func( row, x ), axis = 1 )
display( df )
```
