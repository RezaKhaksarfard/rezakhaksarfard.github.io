# Sort dataframe column by the order of a list using pandas.Categorical

```python
import pandas as pd

df = pd.DataFrame( { 'A': [ 'a', 'b', 'a', 'a', 'b', 'c', 'e', 'd'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )

display( df )
```

```python
df[ 'category' ] = pd.Categorical( df[ 'A' ], ordered = True, categories = [ 'c', 'a', 'd', 'e', 'b' ] )
df.sort_values( 'category', inplace = True )
df.drop( [ 'category' ], axis = 1, inplace = True )
display( df )
```
