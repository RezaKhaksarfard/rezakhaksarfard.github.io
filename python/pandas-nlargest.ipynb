# Filter dataframe groups by largest size applying pandas.Series.nlargest

```python
import pandas as pd

df = pd.DataFrame( { 'A': [ 'a', 'b', 'a', 'a', 'b', 'c', 'e', 'd'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )

display( df )
```

```python
indexList = df.groupby( 'A' ).size().nlargest( 2 ).index.tolist() 
print( 'Index List = ', indexList )

display( df.loc[ df[ 'A' ].isin( indexList ) ] )
```
