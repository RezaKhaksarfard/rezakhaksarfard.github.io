# Add new column to dataframe groups by pandas.core.groupby.GroupBy.apply

```python
import pandas as pd

df = pd.DataFrame( { 'A': [ 'a', 'a', 'b', 'a', 'b', 'b', 'a', 'b'],
                     'B': [  1 ,  4 ,  2 ,  5 ,  5 ,  6 ,  3 ,  1 ],
                     'C': [  9 ,  7 ,  3 ,  5 ,  1 ,  4 ,  7 ,  2 ] } )

display( df )
```

```python
def func( group ):
    group[ 'C_CumSum' ] = group[ 'C' ].cumsum()
    return group

display( df.groupby( by = 'A' ).apply( func ).sort_values( by = 'A').reset_index( drop = True ) )
```
